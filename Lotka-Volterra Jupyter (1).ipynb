# Modelo Lotka-Volterra

## ¿Que es?

El modelo Lotka-Volterra, también conocido como el modelo depredador-presa, es un conjunto de ecuaciones diferenciales que describe la dinámica entre dos especies biológicas que interactúan, típicamente un depredador y una presa. Fue desarrollado independientemente por Alfred Lotka en 1925 y Vito Volterra en 1926.

Se importan 3 librerias las cuales son:

Matplotlib (matplotlib.pyplot as plt) para crear gráficos.

Random (random) para generar números aleatorios.

Numpy (numpy) para realizar cálculos matemáticos y manipular matrices.

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys
```

Con estos parámetros, se construye y ejecuta una simulación del modelo Lotka-Volterra

Se definen las poblaciones iniciales de la presa (x) y del depredador (y).
Se establecen los parámetros como las tasas de crecimiento, interacción, y decrecimiento para modelar la dinámica de las poblaciones.
Simulación:

Se utiliza un bucle de tiempo que avanza en pasos pequeños (definidos por dt) desde 0 hasta max_time.
En cada paso de tiempo, se calculan las nuevas poblaciones de presa y depredador utilizando las ecuaciones diferenciales que dependen de los parámetros a, b, c, y e.

a = 0.7: Tasa de crecimiento de la población presa (x).
    
b = 0.5: Tasa de interacción negativa entre la presa (x) y el depredador (y).
    
c = 0.3: Tasa de decrecimiento de la población depredadora (y).
    
e = 0.2: Tasa de interacción positiva entre el depredador (y) y la presa (x).
    
dt = 0.001: Paso de tiempo para la simulación.
    
max_time = 100: Tiempo máximo de la simulación.
    
Estos parámetros se utilizan para definir la dinámica entre las poblaciones de presas y depredadores en el modelo Lotka-Volterra.

```python
# Parámetros del modelo
a = 0.7  # Tasa de crecimiento de la población x
b = 0.5  # Tasa de interacción negativa entre las poblaciones x e y
c = 0.3  # Tasa de decrecimiento de la población y
e = 0.2  # Tasa de interacción positiva entre las poblaciones x e y
dt = 0.001  # Paso de tiempo
max_time = 100  # Tiempo máximo de simulación
```

Tiempo inicial (t = 0):

Se establece el punto de partida temporal de la simulación. Esto significa que la simulación comienza en el tiempo t = 0.
Población inicial de la presa (x = 3.0):

Se define cuántas unidades de la especie presa (x) hay al inicio de la simulación. Aquí, se comienza con 3.0 presas.
Población inicial del depredador (y = 0.5):

Se establece cuántas unidades de la especie depredadora (y) hay al inicio de la simulación. En este caso, se comienza con 0.5 depredadores.


```python
# Inicialización del tiempo y las poblaciones
t = 0   # Tiempo inicial
x = 3.0  # Población inicial de x
y = 0.5  # Población inicial de y
```

En esta parte del código, se crean listas vacías para almacenar los valores generados durante la simulación. Aquí te explico qué se hace:

t_list = []:

Se crea una lista vacía llamada t_list que almacenará los valores del tiempo en cada paso de la simulación.
x_list = []:

Se crea una lista vacía llamada x_list para almacenar los valores de la población de la presa (x) en cada paso de la simulación.
y_list = []:

Se crea una lista vacía llamada y_list que guardará los valores de la población del depredador (y) en cada paso de la simulación

```python
# Listas vacías para almacenar el tiempo y las poblaciones
t_list = []
x_list = []
y_list = []
```

En esta parte del código, se inicializan las listas creadas anteriormente con los valores iniciales de tiempo y poblaciones. Específicamente:

t_list.append(t):

Agrega el valor actual de t (que es 0) a la lista t_list. Esto registra el tiempo inicial de la simulación.
x_list.append(x):

Agrega el valor actual de x (la población inicial de la presa, que es 3.0) a la lista x_list. Esto registra la población inicial de la presa.
y_list.append(y):

Agrega el valor actual de y (la población inicial del depredador, que es 0.5) a la lista y_list. Esto registra la población inicial del depredado

```python
# Inicializar las listas con los valores iniciales
t_list.append(t)
x_list.append(x)
y_list.append(y)
```

En esta parte del código, se realiza el bucle principal de la simulación y se grafican los resultados

while t < max_time::

Se ejecuta el bucle mientras el tiempo t sea menor que el tiempo máximo max_time.

t = t + dt:

Incrementa el tiempo t en el paso de tiempo dt, avanzando la simulación en pequeños intervalos.

x = x + (a * x - b * x * y) * dt:

Calcula el nuevo valor de la población de la presa x utilizando la ecuación del modelo, que tiene en cuenta el crecimiento natural de la presa y la reducción causada por la interacción con los depredadores.

y = y + (-c * y + e * x * y) * dt:

Calcula el nuevo valor de la población del depredador y usando la ecuación del modelo, considerando la disminución natural de los depredadores y el aumento debido a la interacción con las presas.

```python
while t < max_time:
    # Calcular nuevos valores de t, x, y
    t = t + dt
    x = x + (a * x - b * x * y) * dt
    y = y + (-c * y + e * x * y) * dt
```

Se están almacenando los valores actualizados de tiempo y poblaciones en las listas correspondientes. Aquí está el detalle:

t_list.append(t):

Añade el valor actualizado de t (el tiempo actual en la simulación) a la lista t_list. Esto permite rastrear cómo cambia el tiempo a medida que avanza la simulación.

x_list.append(x):

Añade el valor actualizado de x (la población de presas en el tiempo actual) a la lista x_list. Esto permite seguir la evolución de la población de presas.

y_list.append(y):

Añade el valor actualizado de y (la población de depredadores en el tiempo actual) a la lista y_list. Esto permite seguir la evolución de la población de depredadores

```python
 # Almacenar los nuevos valores en las listas
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)
```

Se está utilizando la biblioteca Matplotlib para crear y mostrar una gráfica con los resultados de la simulación.

p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth=2):

plt.plot(): Crea un gráfico de líneas con los datos proporcionados.

t_list, x_list, 'r': Grafica los datos de tiempo (t_list) en el eje x y la población de presas (x_list) en el eje y con una línea de color rojo ('r').

t_list, y_list, 'g': Grafica los datos de tiempo (t_list) en el eje x y la población de depredadores (y_list) en el eje y con una línea de color verde ('g').

linewidth=2: Establece el ancho de las líneas en el gráfico a 2 para que sean más visibles.
plt.show():

Muestra la gráfica generada en una ventana emergente. Es el paso final para visualizar el gráfico que has creado con los datos de la simulación.

```python
# Graficar los resultados
p = plt.plot(t_list, x_list, 'r',  # Graficar población x en rojo
             t_list, y_list, 'g',  # Graficar población y en verde
             linewidth=2)  # Ancho de línea
plt.show()  # Mostrar la gráfica


```

# PRIMER MODIFICACION


## Se aumentará el valor de presas que en este código es la letra x agregándole el valor de 3.0 remplazando el valor principal que era 1.0 en base a esto el número de depredadores al inicio aumentara debido a que hay más presas disponibles.


```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys

# Parámetros del modelo
a = 0.7  # Tasa de crecimiento de la población x
b = 0.5  # Tasa de interacción negativa entre las poblaciones x e y
c = 0.3  # Tasa de decrecimiento de la población y
e = 0.2  # Tasa de interacción positiva entre las poblaciones x e y
dt = 0.001  # Paso de tiempo
max_time = 100  # Tiempo máximo de simulación

# Inicialización del tiempo y las poblaciones
t = 0   # Tiempo inicial
x = 3.0  # Población inicial de x
y = 0.5  # Población inicial de y

# Listas vacías para almacenar el tiempo y las poblaciones
t_list = []
x_list = []
y_list = []

# Inicializar las listas con los valores iniciales
t_list.append(t)
x_list.append(x)
y_list.append(y)

# Bucle principal de simulación
while t < max_time:
    # Calcular nuevos valores de t, x, y
    t = t + dt
    x = x + (a * x - b * x * y) * dt
    y = y + (-c * y + e * x * y) * dt

    # Almacenar los nuevos valores en las listas
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Graficar los resultados
p = plt.plot(t_list, x_list, 'r',  # Graficar población x en rojo
             t_list, y_list, 'g',  # Graficar población y en verde
             linewidth=2)  # Ancho de línea

plt.show()  # Mostrar la gráfica
```

# SEGUNDA MODIFICACION:

En esta modificación se cambiará el número de depredadores que es la letra y a 3.5 sustituyendo el 0.5 original en base a eso habrá más presión de caza para las presas que podría bajar o disminuir más su población, debido a eso no habrá mucho a alimento y disminuirá su población con el tiempo.


```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys

# Parámetros del modelo
a = 0.7; b = 0.5; c = 0.3;  e = 0.2
dt = 0.001; max_time = 100

# Inicialización del tiempo y las poblaciones
t = 0; x = 1.0; y = 3.5

#Listas Vacias 

t_list = []; x_list = []; y_list = []

# initialize lists
t_list.append(t); x_list.append(x); y_list.append(y)
while t < max_time:
    # calc new values for t, x, y
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt
    # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)
# Plot the results    
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
plt.show()
```
